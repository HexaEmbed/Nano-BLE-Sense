```python
import tensorflow as tf
import numpy as np

# Training data
X = np.array([1.3, 2.7, 3.3, 4.9, 5.4, 6.3, 7.9, 8.3, 9.1, 10.99], dtype=np.float32)
y = 2 * X

# Define a simple model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X, y, epochs=2300, verbose=0)

# Test
usr_input = float(input("Enter x to predict y=2x: "))
print("Prediction for x=10:", model.predict(np.array([usr_input])))
```

```python
# Convert to TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save to file
with open("model.tflite", "wb") as f:
    f.write(tflite_model)
```
