```python
# Install scikit-learn if not already available
!pip install -q scikit-learn

# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Create training data
# X must be 2D for sklearn, so we reshape
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = 2 * X

# Define and train the model
model = LinearRegression()
model.fit(X, y)

# Check learned parameters
print("Coefficient (slope):", model.coef_[0][0])
print("Intercept:", model.intercept_[0])

# Test prediction
test_value = np.array([[10]])
prediction = model.predict(test_value)
print("Prediction for x=10:", prediction[0][0])
```
